```python
def decToEnglish(baseTenNum):
  hundreds = 0
  tens = 0
  ones = 0
  englishRepresentation = ""
  #This variable is used to avoid having a number like 12 be output as "Twelve Zero"
  uniqueNumber = False
  #Segment baseTenNum into parts. Keep in mind that numbers like 12 are read
  #"Twelve" and not "Ten two"
  #If baseTenNum % 100 is between 11 and 19, do not assign anything to ones
  if baseTenNum % 100 >= 10 and baseTenNum % 100 <= 19:
    uniqueNumber = True
    tens = baseTenNum % 100
    baseTenNum -= tens
    hundreds = baseTenNum % 1000
    baseTenNum -= hundreds
  #else, it's safe to assume there's no number between 11 and 19 in the argument
  else:
    ones = baseTenNum % 10
    baseTenNum -= ones
    tens = baseTenNum % 100
    baseTenNum -= tens
    hundreds = baseTenNum % 1000
    baseTenNum -= hundreds
  
  #Begin adding English representations, starting with hundreds
  match(hundreds):
    case 100:
      englishRepresentation += "sada  "
    case 200:
      englishRepresentation += "kakssada "
    case 300:
      englishRepresentation += "kolmsada "
    case 400:
      englishRepresentation += "nelisada "
    case 500:
      englishRepresentation += "viissada "
    case 600:
      englishRepresentation += "kuussada "
    case 700:
      englishRepresentation += "seitsesada "
    case 800:
      englishRepresentation += "kaheksasada "
    case 900:
      englishRepresentation += "üheksasada "
  #Next English representations, add tens
  match(tens):
    case 10:
      englishRepresentation += "kümme "
    case 11:
      englishRepresentation += "üksteist "
    case 12:
      englishRepresentation += "kaksteist "
    case 13:
      englishRepresentation += "kolmteist "
    case 14:
      englishRepresentation += "neliteist "
    case 15:
      englishRepresentation += "viisteist "
    case 16:
      englishRepresentation += "kuusteist "
    case 17:
      englishRepresentation += "seitseteist "
    case 18:
      englishRepresentation += "kaheksateist "
    case 19:
      englishRepresentation += "üheksateist "
    case 20:
      englishRepresentation += "kakskümmend "
    case 30:
      englishRepresentation += "kolmkümmend "
    case 40:
      englishRepresentation += "nelikümmend "
    case 50:
      englishRepresentation += "viiskümmend "
    case 60:
      englishRepresentation += "kuuskümmend "
    case 70:
      englishRepresentation += "seitsekümmend "
    case 80:
      englishRepresentation += "kaheksakümmend "
    case 90:
      englishRepresentation += "üheksakümmend "
  #Ones match should only happen if uniqueNumber = false to avoid a number
  #like 12 being read as "Twelve Zero"
  if uniqueNumber == False:
    match(ones):
      case 0:
        englishRepresentation += "null"
      case 1:
        englishRepresentation += "üks"
      case 2:
        englishRepresentation += "kaks"
      case 3:
        englishRepresentation += "kolm"
      case 4:
        englishRepresentation += "neli"
      case 5:
        englishRepresentation += "viis"
      case 6:
        englishRepresentation += "kuus"
      case 7:
        englishRepresentation += "seitse"
      case 8:
        englishRepresentation += "kaheksa"
      case 9:
        englishRepresentation += "üheksa"
  #Remove any trailing spaces and return english representation
  #Lesson learned: the rstrip() method can remove trailing characters. Space is
  #the default character
  return(englishRepresentation.rstrip())

print(decToEnglish(765))
```
